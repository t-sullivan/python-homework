```python
'''Your task is to create a Python script that analyzes the records to calculate each of the following:

The total number of months included in the dataset

The net total amount of Profit/Losses over the entire period

The average of the changes in Profit/Losses over the entire period

The greatest increase in profits (date and amount) over the entire period

The greatest decrease in losses (date and amount) over the entire period'''

from pathlib import Path
import csv

csvpath = "C:/Users/tlsul/Desktop/bootcamp/PyBank/Resources/budget_data.csv"
Total_Months = 0
net_profit = 0
max_profit = 0
min_profit = 0
profit_by_month = {}
Total_Months = []
with open(csvpath, 'r') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    header = next(csvreader)
        
    #date and profit    
    for row in csvreader:
        date = row[0]
        profit_by_month[date] = int(row[1])
        
        
Total_Months = len(profit_by_month.keys())

previous_mon = 0
current_mon = 0
avg_change = 0
max_gain = 0
max_loss = 10000000000
max_month =""
min_month =""

for key in profit_by_month:
    current_mon = profit_by_month[key]
    net_profit += current_mon
    
    if previous_mon != 0:
        difference = current_mon - previous_mon
        avg_change += difference
        
        if difference > max_gain:
            max_gain = difference
            max_month = key
            
        if difference < max_loss:
            max_loss = difference
            min_month = key
  
    previous_mon = current_mon
    
net_profit = round(net_profit,2)
average = round(avg_change/(Total_Months-1),2)

print("")
print("------------------")
print("Financial Analysis")
print("------------------")
print("")
print(f"Total Months: {Total_Months}") 
print(f"Net Profit: ${net_profit}")
print(f"Average Change: ${average}")
print(f'Greatest Increase in Profits: {max_month} ${max_gain}')
print(f'Greatest Decrease in Profits: {min_month} ${max_loss}')


# Set output file name
output_path = 'homework1.txt'

# Open the output path as a file object
with open(output_path, 'w') as file:
    file.write("")
    file.write("------------------")
    file.write("Financial Analysis")
    file.write("------------------")
    file.write("")
    file.write(f"Total Months: {Total_Months}") 
    file.write("|")
    file.write(f"Net Profit: ${net_profit} ")
    file.write("|")
    file.write(f"Average Change: ${average}")
    file.write("|")
    file.write(f'Greatest Increase in Profits: {max_month} ${max_gain}')
    file.write("|")
    file.write(f'Greatest Decrease in Profits: {min_month} ${max_loss}')
```
